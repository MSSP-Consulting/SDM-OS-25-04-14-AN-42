```python
import pandas as pd


df1 = pd.read_excel('data set research.xlsx', sheet_name='BEFORE HH', header=2)

df2 = pd.read_excel('data set research.xlsx', sheet_name='AFTER HH', header=2)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 假设 df1 = BEFORE, df2 = AFTER
df1['Group'] = 'Before'
df2['Group'] = 'After'

# 合并数据
df = pd.concat([df1, df2], ignore_index=True)

# 去除列名中多余空格
df.columns = df.columns.str.strip()
```

```python
df.to_csv('data.csv', index=False)
df
```

```python

plt.figure(figsize=(8, 6))
sns.boxplot(x='Group', y='LOS (days)', data=df)
plt.title('LOS (days) Before vs After')
plt.show()
```

```python
sns.countplot(x='Total Intravenous Anesthesia (TIVA)  vs. Inhalational Gas Anesthesia', hue='Group', data=df)
plt.xticks(rotation=45)
plt.title('Anesthesia Type Before vs After')
plt.show()
```

```python

plt.figure(figsize=(16, 5))

features = ['INTRA-OPERATIVE IV FLUIDS (mL)', 'EBL (mL)', 'SURGICAL TIME (min)']
for i, feature in enumerate(features):
    plt.subplot(1, 3, i+1)
    sns.boxplot(x='Group', y=feature, data=df)
    plt.title(f'{feature}: Before vs After')
    plt.xticks(rotation=15)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
df['CO-MORBIDITIES'] = df['CO-MORBIDITIES'].str.lower()
df['PROCEDURE'] = df['PROCEDURE'].str.lower()
plt.figure(figsize=(18, 12))

# 连续变量：AGE & BMI
for i, feature in enumerate(['AGE', 'BMI']):
    plt.subplot(3, 2, i + 1)
    sns.histplot(data=df, x=feature, hue='Group', kde=True, element='step', stat='density', common_norm=False)
    plt.title(f'{feature} Distribution by Group')

# 类别变量：GENDER, CO-MORBIDITIES, SMOKER?, PROCEDURE
categorical_features = ['GENDER', 'CO-MORBIDITIES', 'SMOKER?', 'PROCEDURE']
for i, feature in enumerate(categorical_features):
    plt.subplot(3, 2, i + 3)
    sns.countplot(data=df, x=feature, hue='Group')
    plt.title(f'{feature} Count by Group')
    plt.xticks(rotation=30)

plt.tight_layout()
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 定义连续型变量
fluid_features = [
    'INTRA-OPERATIVE IV FLUIDS (mL)',
    'PO intake (mL) on POD 1',
    'PO intake (mL) on POD 2 (if applicable)',
    'POST-OP IV Fluids (mL)'
]

plt.figure(figsize=(18, 10))

for i, feature in enumerate(fluid_features):
    plt.subplot(2, 2, i+1)
    sns.boxplot(data=df, x='Group', y=feature)
    plt.title(f'{feature}: Before vs After')
    plt.xticks(rotation=15)

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

sns.scatterplot(
    data=df,
    x='BMI',
    y='INTRA-OPERATIVE IV FLUIDS (mL)',
    hue='Group',
    style='Group',  # 可以加上不同形状区分 Before/After
    alpha=0.7,
    s=70  # 点大小
)

plt.title('BMI vs Intra-Op IV Fluids by Group')
plt.xlabel('BMI')
plt.ylabel('Intra-Operative IV Fluids (mL)')
plt.legend(title='Group', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 创建一个新列：是否有并发症备注
df['Complication Notes'] = df['OTHER NOTES/COMPLICATIONS (infections/delayed wound healing, etc.)'].apply(
    lambda x: 'yes' if pd.notnull(x) and str(x).strip() != '' else 'no'
)

# 计算占比
complication_counts = df.groupby(['Group', 'Complication Notes']).size().reset_index(name='Count')

# 绘图
plt.figure(figsize=(8, 6))
sns.barplot(
    data=complication_counts,
    x='Group',
    y='Count',
    hue='Complication Notes'
)

plt.title('Complication Notes:Before vs After')
plt.ylabel('population')
plt.xlabel('Group')
plt.legend(title='Complication')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot 1: Opioid Use vs Emesis Events (Barplot)
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='FLOOR OPIOID USE', y='# EMESIS EVENTS', hue='Group')
plt.title('Emesis Events vs Opioid Use by Group')
plt.xlabel('Opioid Use')
plt.ylabel('# Emesis Events')
plt.legend(title='Group', bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()

# Plot 2: Anti-Emetic Use vs Emesis Events (Barplot)
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='FLOOR ANTI-EMETICS', y='# EMESIS EVENTS', hue='Group')
plt.title('Emesis Events vs Anti-Emetic Use by Group')
plt.xlabel('Anti-Emetic Use')
plt.ylabel('# Emesis Events')
plt.legend(title='Group', bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```

```python
# 做一个交叉表（contingency table）
cross_tab = pd.crosstab(
    [df['Group'], df['FLOOR OPIOID USE']],
    df['FLOOR ANTI-EMETICS']
)

plt.figure(figsize=(8, 5))
sns.heatmap(cross_tab, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Cross-Tab: Opioid Use vs Anti-Emetics by Group')
plt.tight_layout()
plt.show()
```
